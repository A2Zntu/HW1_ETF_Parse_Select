# Setting

* 使用BeautifulSoup進行html解析
* 使用requests發送網路請求
* 使用webbrowser模擬點擊url
* 使用urllib.parse解析網址

```python
from bs4 import BeautifulSoup
import requests
import webbrowser
from urllib.parse import urlparse
import pandas as pd
```

# 取得各基金home page

```python
data = pd.read_csv("group17_region.csv") 
data['Home page']=""
for s in data['Symbol']:
    print(s)
    etfdbpage = "https://etfdb.com/etf/"+s
    print(etfdbpage)
    page_response = requests.get(etfdbpage, timeout=5)
    page_content = BeautifulSoup(page_response.content, "html.parser")
    test = page_content.find_all("a", string="Home page")
    testresult = str(test).split('"')[1]
    if '&amp;' in testresult:
        testresult = testresult.replace("&amp;","&")
    print(testresult)
    data.loc[data['Symbol']==s, 'Home page']=testresult
    
```

```python
data.to_csv('website.csv')
```

# iShares

```python
ishares = data[data['Group']=='iShares']
ishares
```

```python
urllist = list(ishares['Home page'])
urllist
# page_link ='https://www.ishares.com/us/products/239601/?referrer=tickerSearch'
```

```python
for url in urllist:
    page_link = url
    page_response = requests.get(page_link, timeout=5, allow_redirects=True)
    page_link = page_response.url
    parsed_uri = urlparse(page_link)
    result = '{uri.scheme}://{uri.netloc}/'.format(uri=parsed_uri)
    print(result)
    page_content = BeautifulSoup(page_response.content, "html.parser")
    test = page_content.find_all(class_='icon-xls-export')
    for tags in test:
        if "fund download:common" in str(tags):
            print(tags.get('href'))
            therest = tags.get('href')
            thelink = result + therest
            webbrowser.open(thelink)
    
```
